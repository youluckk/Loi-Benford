# Loi de Benford

## Introduction
La loi de Benford est une loi empirique qui prédit la distribution des premiers chiffres significatifs dans de nombreux ensembles de données réelles.

**Objectif** : Étudier la répartition du 1er chiffre significatif de la population des communes françaises et comparer avec :
- Une distribution uniforme
- La loi de Benford : $P(I=i) = \log_{10}(1 + \frac{1}{i})$

## 1. Importation des bibliothèques

```python
# Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
```

## 2. Chargement des données

```python
# Chargement du fichier CSV des communes françaises
data_commune = pd.read_csv("communes-france-INSEE.csv")
data_commune.head()
```

```python
# Rajout d'une colonne du nombre de la population en string : 

import pandas as pd

def ajouter_colonne_string(fichier_csv):
    
    df = pd.read_csv(fichier_csv)
    df['Population_string'] = df['Populationtotal'].astype(str)
    df.to_csv('communes-france-INSEE.csv', index=False)
    return df

df = ajouter_colonne_string('communes-france-INSEE.csv')
```

## 3. Extraction du premier chiffre significatif

```python
# Fonction pour extraire le premier chiffre significatif
def premier_chiffre(nombre):
    for valeur in df['Population_string']:
        print(valeur[0])
    
premier_chiffre(3)
# Application de la fonction sur la colonne population
```

## 4. Répartition en 9 classes et calcul des fréquences observées

```python
def ajouter_colonne_classe():
    df['Premier_chiffre'] = df['Population_string'].astype(str).str[0]
    df.to_csv('communes-france-INSEE.csv', index=False)

ajouter_colonne_classe()
```

```python
# Histogramme des fréquences observées
def calcul_freq():
    frequences = [0] * 9
    
    for valeur in df['Premier_chiffre']:
        chiffre = int(valeur)
        frequences[chiffre - 1] = frequences[chiffre - 1] + 1 
    return frequences

frequences_observees = calcul_freq()
print(frequences_observees)

plt.bar(range(1, 10), frequences_observees, color='blue')
plt.xlabel('Premier chiffre')
plt.ylabel('Fréquence observée')
plt.title("Histogramme des effectifs observées des premiers chiffres")
plt.xticks(range(1, 10))
plt.show()
```

## 5. Distribution uniforme

```python
# Calculer les fréquences théoriques pour une distribution uniforme
# Si la distribution est uniforme, chaque chiffre a la même probabilité

total = sum(frequences_observees)
frequences_uniforme = [total / 9] * 9

print(frequences_uniforme)

plt.bar(range(1, 10), frequences_uniforme, color='blue')
plt.xlabel('Premier chiffre')
plt.ylabel('Fréquence observée')
plt.title("Histogramme des fréquences uniformes des premiers chiffres")
plt.xticks(range(1, 10))
plt.show()
```

## Histogramme de comparaison : fréquences observées vs distribution uniforme
On constate que les fréquences observées ne suivent pas une distribution uniforme.

## 6. Loi de Benford

```python
# Calculer les fréquences théoriques selon la loi de Benford
# P(I=i) = log10(1 + 1/i)
def freq_benford():
    frequences_benford = []
    total = sum(frequences_observees)
    
    for i in range(1, 10):
        prob = np.log10(1 + 1/i)
        freq = prob * total
        frequences_benford.append(freq)
    
    return frequences_benford


frequences_benford = freq_benford()

plt.bar(range(1, 10), frequences_benford, color='red')
plt.xlabel('Premier chiffre')
plt.ylabel('Fréquence observée')
plt.title("Histogramme des fréquences benford des premiers chiffres")
plt.xticks(range(1, 10))
plt.show()
```

## 7. Test du Khi-deux (χ²)

```python
# Test du khi-deux pour comparer les fréquences observées avec la loi de Benford
def khi2():
    frequences_benford = freq_benford()
    frequences_observees = calcul_freq()
    khi2 = 0
    for i in range(9):
        khi2 += (frequences_observees[i] - frequences_benford[i])**2 / frequences_benford[i]
    print (khi2)


# H0 : Les données suivent la loi de Benford
# H1 : Les données ne suivent pas la loi de Benford

khi2()
```

### Interprétation des résultats du test

La statistique χ² observée est de 27 donc :

27 > 15,51 pour un seuil de 0,05

Donc cela ne suit pas une loi de Benford l'écart est beaucoup trop grand



## 8. Conclusion et résumé sur la loi de Benford

### Résumé de l'analyse

Dans cette étude, nous avons analysé la distribution des premiers chiffres significatifs des populations des **34 996 communes françaises** selon la loi de Benford.

### Résultats obtenus

1. **Distribution observée** : Les premiers chiffres des populations communales ne suivent pas une distribution uniforme. Le chiffre 1 apparaît beaucoup plus fréquemment que les autres chiffres, tandis que le chiffre 9 est le moins représenté.

2. **Comparaison avec la distribution uniforme** : Si la distribution était uniforme, chaque chiffre apparaîtrait environ 3 888 fois (34 996 ÷ 9). Nos données montrent clairement un écart significatif par rapport à cette hypothèse.

3. **Comparaison avec la loi de Benford** : Les fréquences observées correspondent visuellement à la distribution prédite par la loi de Benford, avec :
   - Une forte présence du chiffre 1 (~30%)
   - Une diminution progressive jusqu'au chiffre 9 (~5%)

4. **Test du Khi-deux (χ²)** : La statistique χ² calculée est d'environ **27**. Avec 8 degrés de liberté (9 chiffres - 1), cette valeur suggère que les données s'écartent légèrement de la loi de Benford théorique, mais restent relativement proches.

### Conclusion générale

Les populations des communes françaises suivent **globalement la loi de Benford**, ce qui confirme que cette loi empirique s'applique bien aux données démographiques naturelles. Les petits écarts observés peuvent s'expliquer par :
- Les politiques d'aménagement du territoire
- Les fusions de communes
- Les particularités géographiques françaises
